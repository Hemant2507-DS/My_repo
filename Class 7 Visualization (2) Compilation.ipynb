```python
students = ['Ramesh', 'Suresh','Adesh', 'Jignesh','Mahesh']
salary = [10000, 35000, 18000, 44000, 23000]
salary2 = [20000, 45000,30000,60000,44000]
```

```python
plt.figure(figsize = (6,4)) # 8 is width, 6 is height
plt.plot(students, salary, linestyle = '--', linewidth = 1, marker = '+', markerfacecolor = 'b')
plt.plot(students, salary2, linestyle = '--', linewidth = 1, marker = '.', markerfacecolor = 'r')
plt.xlabel('students')
plt.ylabel('salary')
plt.grid(True, linestyle = '--')
plt.legend(['salary','salary2']) # To separate 2 entries in legend use a list[]
plt.title('Previous year Vs This Year Salary')
plt.show()
```

### Popular plots in Matplotlib
1. Line Plot (plt.plot)
2. Bar plot (plt.bar) or (plt.barh)
3. Scatter Plot (plt.scatter)
4. Histogram (plt.hist) Imp to remember edgecolor = ''
5. Area Plot: Use stackplot(plt.stackplot)
6. Pie Plot (plt.pie) Imp to remember(labels = '', autopct = '%1.1f', explode = explode)
7. Box Plot (plt.boxplot)

You can also create a number of plots inside a quarter as 'Subplots', e.g. one plot can be pie chart, other can be a line plot, scatter plot, barplot and more.

```python
x = ['TV', 'Radio', 'Internet']
y = [[30,45,23],[25,55,7]]
```

### Plotting a Scatter Plot

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
Temperature_1_week = [29,26,35,40,38]
Temperature_2_week = [34,31,37,42,47]
plt.scatter(days, Temperature_1_week, marker = '.', color = 'b')
plt.scatter(days, Temperature_2_week, marker = '+', color = 'r')
plt.xlabel('Days')
plt.ylabel('Tempearture')
plt.title('Weather Forecast Change')
plt.grid(True, linestyle = '--')
plt.legend(['First Week', 'Second Week'])
plt.show()
```

### Box Plot and Violin plot 
Both have the same functionality -> used in outlier analysis.
1. Middle or median  = 50th Percentile
2. lower Percentile = 25th Percentile
3. Upper Percentile = 75th Percentile

Consider num = [19,61,45,96,31,78,98,107,221]
1. Median = 31
2. 25th Percentile  = 61+45/2
3. 75th Percentile = 98+107/2
4. Min = 19
5. Max = 221

```python
marks = np.random.randint(33,68,20)
marks
```

```python
num = [19,61,45,96,31,78,98,107,221]
plt.boxplot(num)
plt.show()
```

```python
num = [19,61,45,96,31,78,98,107,221]
plt.violinplot(num)
plt.show()
```

### Plotting Pie Charts 


```python
Electronic_appliances = ['TVs', 'Washing Machine', 'Air Conditioners']
Sales_value = [5600000, 8700000, 9800000]
explode = [0,0,0.08]
plt.pie(Sales_value, labels = Electronic_appliances, autopct = '%1.1f', explode = explode) # Instead of 'Y' axis we use 'Labels' here
plt.show()
```

### Plotting Donut chart

```python
department = ['Sales', 'Tech']
Strength = [250, 650]
Subgroup_dept = ['S1','S2','ML','AI','DS']
Subgroup_strength = [100, 150, 200, 200, 150]
plt.pie(Strength, labels = department, radius = 1)
plt.pie(Subgroup_strength, labels = Subgroup_dept, radius = 1-0.3, labeldistance = 0.8)
plt.pie([10],radius = 0.5, colors = 'w')
plt.show()
```
