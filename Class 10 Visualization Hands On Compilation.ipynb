### Q2. Create bar plots for the following, make sure all the plots appear in the same plane with x-labels and y-labels evenly spaced. a. Create a bar plot that shows the visual representation of hp, mpg, weight, time-to-60 with respect to the number of cylinders for each of the brands. b. Create a bar plot that shows the visual representation of hp, mpg, weight and time-to-60 with respect to the years, for each brand.

```python
plt.figure(figsize=(12,11))

plt.subplot(2, 2, 1)
plt.bar(df['cylinders'], df['hp'])
plt.xlabel('Cylinders')  
plt.ylabel('hp')          
plt.title('Cylinder vs HP')

plt.subplot(2, 2, 2)  
plt.bar(df['cylinders'], df['mpg'])
plt.xlabel('Cylinders')  
plt.ylabel('mpg')          
plt.title('Cylinder vs MPG')

plt.subplot(2, 2, 3)
plt.bar(df['cylinders'], df['weightlbs'])
plt.xlabel('Cylinders')  
plt.ylabel('weightlbs')    
plt.title('Cylinder vs WeightLBS')

plt.subplot(2, 2, 4)
plt.bar(df['cylinders'], df['time-to-60'])
plt.xlabel('Cylinders')   
plt.ylabel('timeto60')   
plt.title('Cylinder vs TimeTo60')

plt.show()
```

### Q3 Create pair plots for the entire data to study various patterns in the data. a. Create pair plots with respect to brand, number of cylinders, year, etc.

```python
# Pairplot can be plotted using 'SNS' and has the layout that is same as a 'Heatmap'.
# The main use of it is in determing the relationship of one column with other i.e., if there is high dependency of one columns 
# On another (multi-collinearity) so we can remove them. For example, if 'hp' range is from 100-300 but we increase it to 1000.
# Then, cylinders that show high multicollinearity will also show an increase in numbe though the numbers of cylinders are limited.
# So to avoid unwanted or unnecessary errors we can remove one of the columns.

# Alternatively, you can go for 'Correlation'.

sns.pairplot(df)
plt.show()
```

```python
# We choose hue to categorize the data for example, brand has 3 values (US, Japan, Europe). Hue will help us distinguish 
# Between 3 regions i.e. 'US', 'Europe' and 'Japan' by assigning these regions different color codes.
sns.pairplot(df, hue = 'brand')
plt.show()
```

```python
sns.pairplot(df, hue = 'cylinders')
plt.show()
```

```python
sns.pairplot(df, hue = 'year')
plt.show()
```

### Q4. Create a heatmap for the entire data to study correlation between each of the columns.

```python
# corr = sns.heatmap(df, annot = True, cmap = 'Blue')
# corr
# The above line of code will give error since some of the columns data type is object. To plot a matrix we need columns
# That have their datatype as 'float' and 'intiger' only. To do so, use the line of code given below:
# df_new = df.select_dtypes(include = ['int', 'float']).columns
# 
```

```python
df_new = df.select_dtypes(include = ['int', 'float']).columns
df_new
```

```python
df_new = df[df_new]
df_new
```

```python
sns.heatmap(df_new)
```

```python
corr = df_new.corr()
corr
```

```python
sns.heatmap(corr, annot = True, cmap = 'BuPu', fmt = '.3f')
# annot=True: This argument ensures that the values in the heatmap are displayed as annotations on each cell
# .2f" is a format specifier that tells Python to display floating-point numbers with two decimal places
```
