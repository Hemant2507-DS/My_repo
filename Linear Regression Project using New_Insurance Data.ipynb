#### Model Building

```python
# Target_column
y = df.iloc[:,-1]
```

```python
# Independent Column
# Note: Use 2[] else, you'll get too many indexers error
x = df.loc[:,['children','Claim_Amount','past_consultations','num_of_steps','Hospital_expenditure','age']]
```

```python
# Splitting the variable
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 50)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr
```

```python
lr.fit(x_train, y_train)
```

```python
pred = lr.predict(x_test)
pred
```

```python
# Comparing with Y test 
from sklearn.metrics import *
```

```python
score = r2_score(y_test, pred)
score
```
